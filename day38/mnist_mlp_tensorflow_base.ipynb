```python
import tensorflow as tf
import numpy as np
```

```python
from tensorflow import keras

(train_data, train_target), (test_data, test_target) = keras.datasets.mnist.load_data()
```

```python
print(train_data.shape, train_target.shape)

print(test_data.shape, test_target.shape)
```

```python
tf.test.is_gpu_available()
```

```python

def Net():
  model = keras.Sequential()

  model.add(keras.layers.Input(shape=(28, 28, 1)))
  model.add(keras.layers.Conv2D(64, kernel_size=3, strides=1, padding='same'))
  model.add(keras.layers.Activation('relu'))
  model.add(keras.layers.MaxPool2D(3, strides=1, padding='same'))
  model.add(keras.layers.Conv2D(16, kernel_size=2, strides=1, padding='same'))
  model.add(keras.layers.Activation('relu'))
  model.add(keras.layers.MaxPool2D(3, strides=1, padding='same'))

  model.add(keras.layers.AveragePooling2D(1,1))
  model.add(keras.layers.Flatten())
  model.add(keras.layers.Dropout(0.2))
  model.add(keras.layers.Dense(10, activation='softmax'))
  return model
```

```python
model = Net()
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy')
history = model.fit(train_data, train_target, epochs=10, verbose=1)
```

```python
model.evaluate(test_data, test_target)
```
