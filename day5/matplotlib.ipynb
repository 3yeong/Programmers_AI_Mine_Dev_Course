### 2차 함수 그래프 with plot()

```python
# 리스트를 이용해서 1차 함수 y = x를 그려보면:

plt.plot([0,1,2,3,4])
plt.show()
```

```python
#numpy.array를 이용해서 함수 그래프 그리기

x = np.array([1,2,3,4,5]) #정의역
y = np.array([1,4,9,16,25]) #f(x)

plt.plot(x, y)
plt.show()
```

```python
#np.arange(a,b,c) c는 int로 고정되어있다. c를 작게 만들어주기 위해 
#위의 이미지가 너무 딱딱하기 때문에 무수히 많은 점으로 연결시키기

x = np.arange(-20,20,0.01)
plt.plot(x,x**2)
plt.show()
```

```python
#x,y축에 설명 추가하기

x = np.arange(-20,20,0.01)

plt.xlabel("x vlaue")
plt.ylabel("y value")

plt.plot(x,x**2)
plt.show()
```

```python
##,x,y의 범위 설정

x = np.arange(-10,10,0.01)

plt.xlabel("x vlaue")
plt.ylabel("y value")

plt.axis([-5,5,0,25]) #축, 리스트 넣을 수 있다. [x_min, x_max, y_min, y_max]

plt.plot(x,x**2)
plt.show()
```

```python
## x,y축 눈금 설정하기

x = np.arange(-10,10,0.01)

plt.xlabel("x vlaue")
plt.ylabel("y value")

plt.axis([-5,5,0,25]) #축, 리스트 넣을 수 있다. [x_min, x_max, y_min, y_max]

plt.xticks([i for i in range(-5,6,1)]) #눈금을 어디에 설정할지
plt.yticks([i for i in range(0,27,3)])

plt.plot(x,x**2)
plt.show()
```

```python
#그래프에 title 달기

x = np.arange(-10,10,0.01)

plt.xlabel("x vlaue")
plt.ylabel("y value")

plt.axis([-5,5,0,25]) #축, 리스트 넣을 수 있다. [x_min, x_max, y_min, y_max]

plt.xticks([i for i in range(-5,6,1)]) #눈금을 어디에 설정할지
plt.yticks([i for i in range(0,27,3)])

plt.title("y = x^2 graph")

plt.plot(x,x**2)
plt.show()
```

```python
x = np.arange(-10,10,0.01)

plt.xlabel("x vlaue")
plt.ylabel("y value")

plt.axis([-5,5,0,25]) #축, 리스트 넣을 수 있다. [x_min, x_max, y_min, y_max]

plt.xticks([i for i in range(-5,6,1)]) #눈금을 어디에 설정할지
plt.yticks([i for i in range(0,27,3)])

plt.title("y = x^2 graph")

plt.plot(x,x**2, label ="trend")
plt.legend()
plt.show()
```

## 3. Matplotlib Case Study

### 꺽은 선 그래프(Plot)
- `.plot()`

```python
x = np.arange(20)
y = np.random.randint(0,21,20)
x,y
plt.plot(x,y) #plot을 이용하여 꺽은 선 그래프를 그릴 수 있다.

plt.axis([0,20,0,20])

plt.yticks([0,5,10,15,20])


plt.show()

#Extra: y축을 20까지 보이게 하고 싶다면?, y축을 "5"단위로ㅗ 보이게 하고 싶다면?
#.axis(), ,yticks()
```

### 산점도(Scatter Plot)

- `.scatter()`

```python
plt.scatter(x,y) ## 점으로 나타냄
plt.show()
```

 ### 박스 그림 (Box Plot)
 
 - 수치형 데이터에 대한 정보(Q1, Q2,Q3, min, max)

```python
plt.boxplot((x,y))

plt.title("Box plot of x, y")

plt.show()

# Q1 25% , Q2 50%,(median) Q3 75%

#Extra : Plot의 title을 "Box plot of y"
```

### 막대 그래프(Bar Plot)

- 범주형 데이터의 "값"과 그 값의 크기를 직사각형으로 나타낸 그림

```python
plt.bar(x, y)

plt.xticks([i for i in range(0,20,1)])
plt.show()

#Extra : xticks를 올바르게 처리
```

```python
#히스토그램
#도수분포를 직사각형의 막대 형태로 나타낸다
#"계급" : 별량들을 그룹하 하여 나타냄 : 0,1,2가 아니라 0~2까지의 "범주형" 데이터로 구성 후 그림을 그림

plt.hist(y, bins=np.arange(0,22,2))

plt.xticks(np.arange(0,22,2))

plt.show()

#Extra : xticks를 올바르게 고치기
```

### 원형 그래프 (Pie Chart)

- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에 비해서 **비율** 확인에 용이
- `.pie()`

```python
z = [100, 300, 200, 400]

plt.pie(z, labels = ['one', 'two', 'three', 'four'])
plt.show()
```

### 커널밀도그림(Kernel Density Plot)
- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린 그림
- `sns.kdeplot()`

```python
#in histogram

x = np.arange(0,22,2)
y = np.random.randint(0,20,20)

plt.hist(y, bins =x)
plt.show() #y가 도수
```

```python
#kdeplot

sns.kdeplot(y, shade = True)

plt.show() 

#보이게 할 때 공유하므로 써줘야함, 전체를 1로 봤을 때 각부분이 어느정도의 density를 가지는 지 표현
```

### 카운트그림(Count Plot)
- 범주형 column의 빈도수를 시각화 -> Groupby 후의 도수를 하는 것과 동일한 효과
- `sns.countplot()`

```python

vote_df = pd.DataFrame({"name" : ['Andy', "Bob", "Cat"], "vote":[True, True, False]})

vote_df
```

```python
#in matplotlib barplot

vote_count = vote_df.groupby("vote").count()

vote_count
```

```python
plt.bar(x = [False,True], height = vote_count['name'])

plt.show()
```

```python
#sns의 countplot

sns.countplot(x =vote_df['vote'])

plt.show() #label도 있고 구분도 하기 쉽게 되어있음
```

### 캣그림(Cat Plot)
- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- 복잡한 데이터를 구하는데 유리하다
- `sns.catplot()`

```python
covid = pd.read_csv("./country_wise_latest.csv")

covid.head(5)
```

```python
s = sns.catplot(x="WHO Region",y="Confirmed",data=covid)
#, kind='violin'

s.fig.set_size_inches(10,6)

plt.show()

#범주 - who , 범주형 데이터와 수치형 데이터를 맵핑하는데 유리하다.
#여러개의 데이터를 그리는 방법 strip g형태로 보여주는 것이다.
```

### 스트립그림(Strip Plot)
- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프
- 점으로 표현
- `sns.stripplot()`

```python
sns.stripplot(x ='WHO Region' ,y='Recovered', data = covid)
plt.show()
```

```python
#swarmplot 위의 plot과 매우 유사하다
#점이 겹치는 경우 옆으로 분산함

sns.swarmplot(x ='WHO Region' ,y='Recovered', data = covid)
plt.show()
```

### 히트맵(Heatmap)
- 데이터의 행렬을 생상으로 표현해주는 그래프
-상관계수correlation에서 가장 많이 사용한다. corr이 수치로 주어지면 한눈에 알아보기가 어렵다. heatmap을 이용해 시각화 해주면 좋다.
- `sns.heatmap()`

```python
covid.corr()
```

```python
sns.heatmap(covid.corr())

plt.show()
```
